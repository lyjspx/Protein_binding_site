```python
with open("ATPBindingRawFile.txt", "w") as wf1:
    with open("proteinaseBindingRawFile.txt","w") as wf2:
        with open("pdb_seqres.txt", "r") as rf:
            for i in range(0,490318):
                title = rf.readline()
                sequence = rf.readline()
                if title.strip().split(" ")[0][1:] in ATPProteinList.values:
                    #print("catch one")
                    wf1.write(title)
                    wf1.write(sequence)
                if title.strip().split(" ")[0][1:] in proteinaseProteinList.values:
                    #print("catch one")
                    wf2.write(title)
                    wf2.write(sequence)
```
